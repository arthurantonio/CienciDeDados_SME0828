# **5° pergunta**

```python
import requests
from bs4 import BeautifulSoup
```

```python

url = "https://www.worldometers.info/world-population/population-by-country/"
# Send a GET request to the URL
response = requests.get(url)
# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
# Find the table containing the population data
table = soup.find('table', {'id': 'example2'})
# Convert the table to a Pandas DataFrame
df = pd.read_html(str(table))[0]
# Print the DataFrame
df.head()
```

```python
df = df.drop(df.columns[[0, 1]], axis=1)
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
def porcentagem_para_float(valor):
  if(valor[-1] == '%'):
    return float(valor.replace(' %', '')) / 100
  else:
    return np.nan
```

```python
colunasStrings =df.select_dtypes(include='object')

for coluna in colunasStrings:
    df[coluna] = df[coluna].apply(porcentagem_para_float)
```

```python
df = df.dropna()
```

```python
df
```

```python
matriz_correlacao = df.corr()
matriz_correlacao
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Mapa de Calor das Correlações')
plt.show()
```

```python
# Transformar a matriz em um DataFrame e resetar o índice
correlacoes = matriz_correlacao.unstack().reset_index()
correlacoes.columns = ['Variável 1', 'Variável 2', 'Correlação']

# Remover as correlações de uma variável consigo mesma
correlacoes = correlacoes[correlacoes['Variável 1'] != correlacoes['Variável 2']]

# Remover duplicatas (já que a matriz de correlação é simétrica)
correlacoes = correlacoes.drop_duplicates(subset=['Correlação'])

# Selecionar as 5 maiores correlações (em valor absoluto, se necessário)
maiores_correlações = correlacoes.nlargest(5, 'Correlação')

maiores_correlações
```

```python
for index, row in maiores_correlações.iterrows():
    #print(f"A correlação entre {row['Variável 1']} e {row['Variável 2']} é {row['Correlação']:.2f}")

    plt.figure(figsize=(8, 6))
    sns.regplot(x=df[row['Variável 1']], y=df[row['Variável 2']], data=df, scatter_kws={'color': 'blue'})

    # Adicionando título e rótulos
    plt.title('Gráfico de Dispersão com Linha de Tendência')
    plt.xlabel('Coluna A')
    plt.ylabel('Coluna B')

    # Exibindo o gráfico
    plt.show()
```
